### Transposing Arrays and Swapping Axes

Transposing is a special form of reshaping that similarly returns a view on the underlying data without copying anything.

Arrays have the transpose method and also the special T attribute.

```python
arr = np.arange(15).reshape((3, 5))
arr
```

```python
arr.T
```

```python
arr.transpose()
```

```python
arr.T[0] = 0
```

```python
arr
```

```python
arr.T
```

```python
arr = np.random.randn(6, 3)
```

```python
arr
```

```python
np.dot(arr.T, arr)
```

```python
arr = np.arange(16).reshape((2, 2, 4))
```

```python
arr
```

```python
# Here, the axes have been reordered with the second axis first, 
# the first axis second, and the last axis unchanged.
arr.transpose((1, 0, 2))
```

```python
arr
```

```python
# swapaxes similarly returns a view on the data without making a copy.
arr.swapaxes(1, 2)
```

## 4.3 Array-Oriented Programming with Arrays

This practice of replacing explicit loops with array expressions is commonly referred to as ***vectorization***.

```python
points = np.arange(-5, 5, 0.01)
```

```python
xs, ys = np.meshgrid(points, points)
xs
```

```python
ys
```

```python
z = np.sqrt(xs ** 2 + ys ** 2)
```

```python
z
```

### Expressing Conditional Logic as Array Operations

The `numpy.where` function is a vectorized version of the ternary expression `x if con dition else y`.

```python
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])
```

```python
cond = np.array([True, False, True, True, False])
```

```python
result = [x if c else y for (x,y,c) in zip(xarr, yarr, cond)]
result
```

With `np.where` you can write this very concisely:

```python
result = np.where(cond, xarr, yarr)
result
```
